## REGRESSION

**Transformation des valeurs categorielles**

```python
df['school_type'] = df['school_type'].map({'public':0, 'private':1})
df['internet_access'] = df['internet_access'].map({'no':0, 'yes':1})
df['extra_activities'] = df['extra_activities'].map({'no':0, 'yes':1})
```

```python
df['travel_time'] = df['travel_time'].map({'<15 min':0, '15-30 min':1, '30-60 min':2, '>60 min':3})

categorical_cols = ['gender', 'parent_education', 'study_method']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
X = df.drop(columns=['overall_score', 'student_id','final_grade'])
y = df['overall_score']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
X = pd.get_dummies(X, drop_first=True) 

scaler = StandardScaler()

X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)

print(f"MSE: {mse:.2f}")
```

```python
mse = mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse)

print(f"RMSE : {rmse}")
```

**test avec best selection**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import itertools
```

```python
def fit_model(X, y):
    model = LinearRegression()
    model.fit(X, y)
    return model
```

```python
def best_subset_selection(X, y, max_features=6):
    results = []

    for k in range(1, max_features + 1):
        for combo in itertools.combinations(X.columns, k):
            X_subset = X[list(combo)]
            model = fit_model(X_subset, y)
            y_pred = model.predict(X_subset)
            rss = mean_squared_error(y, y_pred) * len(y)

            results.append({
                "n_features": k,
                "features": combo,
                "RSS": rss
            })

    return pd.DataFrame(results)
```

```python
best_models = best_subset_selection(X, y, max_features=6)
```

```python
best_by_k = (
    best_models
    .groupby('n_features')
    .apply(lambda df: df.loc[df['RSS'].idxmin()])
)
```
